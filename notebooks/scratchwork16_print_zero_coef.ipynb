## Simplified String Representation

```python
Zi.unit_strings.reset
```

```python
Zi.include_zero_coefs(True)
```

```python
str(o1)
```

```python
def simple_str_1(z):
    unit_str = Zi.unit_strings.current
    result = ""
    for idx, coef in enumerate(list(utils.flatten(z.to_array()))):
        if idx == 0:
            result += str(coef)
        else:
            result += " + " + str(coef) + unit_str[idx]
    return result
```

```python
simple_str_1(o1)
```

```python
Zi.include_zero_coefs(False)
```

```python
def simple_str_2(z):
    if z.order <= 3:
        unit_str = Zi.unit_strings.current
    else:
        generic_unit_strings = utils.generate_unit_strings(size=z.dim)
        unit_str = Zi.unit_strings.new(generic_unit_strings)
    result = ""
    for idx, coef in enumerate(list(utils.flatten(z.to_array()))):
        if idx == 0:
            result += str(coef)
        else:
            if coef >= 0:
                result += " +" + str(coef) + unit_str[idx]
            else:
                result += " " + str(coef) + unit_str[idx]
    return result
```

```python
simple_str_2(o1)
```

```python
for _ in range(3):
    print(simple_str_2(Zi.random_octonion()))
```

```python
def to_octonion_strings(z):
    result = None
    if z.dim > 8:
        n = z.dim // 8
        if n == 2:
            result = [str(z.real), str(z.imag)]
        else:
            result = [to_octonion_strings(z.real),
                      to_octonion_strings(z.imag)]
    return result
```
