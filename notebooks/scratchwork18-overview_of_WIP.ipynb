## The Construction of Zi's and its Properties

### Zi as a Complex Number (Gaussian Integer)

If the *real* and *imag* parts of the Zi are integers, then the Zi has order 1 and is equivalent to a complex number, or more specifically, a *Gaussian integer*.

```python
print("*** COMPLEX ***\n")

z0 = Zi(3, -4)
print(f"{z0 = }")

print(f"\n{z0.real = }")
print(f"{z0.imag = }")
print(f"{z0.norm = }")
print(f"{abs(z0) = }")
print(f"{z0.order = }  (1 => complex, 2 => quaternion, 3 => octonion, etc.)")
print(f"{z0.dim = }  <-- dim = 2 ** order")
print(f"{z0.conjugate() = }")
print(f"{-z0 = }")
print(f"{complex(z0) = }  <-- NOTE: Only works for Zi of order 1")

print(f"\n{z0.is_complex = }")
print(f"{z0.is_quaternion = }")
print(f"{z0.is_octonion = }")

a, b = z0; print(f"\na, b = z0 --> {a = }, {b = }")

print(f"\n{str(z0) = } <-- NOTE: i is used, rather than j")
print(f"{Zi( '3-4i' ) = }")
print(f"{Zi( '3-4j' ).is_quaternion = } <-- NOTE: because j is a quaternion unit (see example below)")

print(f"\n{z0.to_array() = }")
print(f"{Zi( [3, -4] ) = }")

print(f"\n{Zi() = }")
print(f"{Zi(0) = }")
print(f"{Zi.zero() = }")
print(f"{Zi.one() = }")
print(f"{Zi.eye() = }")
print(f"{Zi.two() = }")
```

### Zi as a Quaternion (Lipschitz Integer)

If the *real* and *imag* parts of a Zi are Zi's of order 1 (complex), then the Zi has order 2 and is equivalent to a quaternion, or more specifically, a *Lipschitz integer*.

```python
print("*** QUATERNION ***\n")

q0 = Zi(Zi(-1, 2), Zi(3, -4))
print(f"{q0 = }")

print(f"\n{q0.real = }")
print(f"{q0.imag = }")
print(f"{q0.norm = }")
print(f"{abs(q0) = }")
print(f"{q0.order = }")
print(f"{q0.dim = }")
print(f"{q0.conjugate() = }")
print(f"{-q0 = }")

print(f"\n{q0.is_complex = }")
print(f"{q0.is_quaternion = }")
print(f"{q0.is_octonion = }")

a, b = q0; print(f"\na, b = q0 --> {a = }, {b = }")

print(f"\n{str(q0) = }")
print(f"{Zi( '-1+2i+3j-4k' ) = }")
print(f"{Zi(    '3-4j'     ) = }  <-- NOTE: From example in previous section")

print(f"\n{q0.to_array() = }")
print(f"{Zi( [[-1, 2], [3, -4]] ) = }")
print(f"{Zi(   [-1, 2, 3, -4]   ) = }")

print(f"\n{Zi.zero(2) = }")
print(f"{Zi.one(2) = }")
print(f"{Zi.eye().increase_order(2) = }")
print(f"{z0 = } => {z0.increase_order(2) = }")
```

### Zi as an Octonion

If the *real* and *imag* parts of a Zi are Zi's of order 2 (quaternion), then the Zi has order 3 and is equivalent to an octonion.

```python
print("*** OCTONION ***\n")

o0 = Zi(Zi(Zi(-1, 2), Zi(3, -4)), Zi(Zi(5, -6), Zi(-7, 8)))
print(f"{o0 = }")

print(f"\n{o0.real = }")
print(f"{o0.imag = }")
print(f"{o0.norm = }")
print(f"{abs(o0) = }")
print(f"{o0.order = }")
print(f"{o0.dim = }")
print(f"{o0.conjugate() = }")
print(f"{-o0 = }")

print(f"\n{o0.is_complex = }")
print(f"{o0.is_quaternion = }")
print(f"{o0.is_octonion = }")

a, b = o0; print(f"\na, b = o0 --> {a = }, {b = }")

print(f"\n{str(o0) = }")
print(f"{Zi( '-1+2i+3j-4k+5L-6I-7J+8K' ) = }")

print(f"\n{o0.to_array() = }")
print(f"{Zi( [[[-1, 2], [3, -4]], [[5, -6], [-7, 8]]] ) = }")
print(f"{Zi( [-1, 2, 3, -4, 5, -6, -7, 8] ) = }")

print(f"\n{Zi.zero(3) = }")
print(f"{Zi.one(3) = }")
print(f"{Zi.eye().increase_order(3) = }")
print(f"{z0 = } => {z0.increase_order(3) = }")
```

## Zi Arithmetic

### Complex Arithmetic

```python
z1 = Zi(10, -7)

print(f"{z0 = }")
print(f"{z1 = }")

print(f"\n{z0 + z1 = }")
print(f"{z0 - z1 = }")
print(f"{z0 * z1 = }")
print(f"NOTE: Using Python's complex numbers: {complex(z0)} * {complex(z1)} = {complex(z0) * complex(z1)}")

print(f"\n{z0 + 3 = }")
print(f"{z0 - 3 = }")
print(f"{z0 * 3 = }")

print(f"\n{3 + z0 = }")
print(f"{3 - z0 = }")
print(f"{3 * z0 = }")
```

### Quaternion Arithmetic

```python
q1 = Zi(Zi(8, -9), Zi(-2, 6))

print(f"{q0 = }")
print(f"{q1 = }")

print(f"\n{q0 + q1 = }")
print(f"{q0 - q1 = }")
print(f"{q0 * q1 = }")
print(f"{q0.hamilton_product(q1) = }")

print(f"\n{q0 + 3 = }")
print(f"{q0 - 3 = }")

print(f"\n{3 + q0 = }")
print(f"{3 - q0 = }")

print(f"\n{q0 + z0 = }")
print(f"{q0 - z0 = }")

print(f"\n{z0 + q0 = }")
print(f"{z0 - q0 = }")
```

With respect to multiplication, if ``use_scalar_mult`` is set to ``True``, then if a lower order Zi is multiplied by a higher order Zi, the multiplication is treated like scalar-vector multiplication, each component of the higher order Zi is multiplied by the lower order Zi.

So, if $z_m$ and $z_n$ are two Zi's where $m$ and $n$ are their respective orders, and $m < n$, then

$z_m \cdot z_n = \text{Zi}(z_m \cdot z_n^\text{real},\ z_m \cdot z_n^\text{imag})$ and

$z_n \cdot z_m = \text{Zi}(z_n^\text{real} \cdot z_m,\ z_n^\text{imag} \cdot z_m)$

Note that both $z_n^\text{real}$ and $z_n^\text{imag}$ will have order n-1. If $m < n - 1$, then the rule will apply recursively.

On the other hand, if ``use_scalar_mult`` is set to ``False``, then the lower order Zi will be "cast" to a Zi with the same order as the higher order Zi, using the method ``increase_order``.

So, $z_m \cdot z_n = (z_m \uparrow n) \cdot z_n$ and

$z_n \cdot z_m = z_n \cdot (z_m \uparrow n)$

This is all illustrated by the examples below.

```python
Zi.use_scalar_mult(True)

print(f"{z0 = }")
print(f"{q1 = }")
print(f"{q1 * 3 = }")
print(f"{q1 * z0 = }")
print(f"{z0 * q1 = }  NOTE: Because complex multiplication is commutative")
print("---- Check using Python complex numbers ----")
print(f"{complex(Zi(3, -4))} * {complex(Zi(8, -9))} = {complex(Zi(3, -4)) * complex(Zi(8, -9))}")
print(f"{complex(Zi(3, -4))} * {complex(Zi(-2, 6))} = {complex(Zi(3, -4)) * complex(Zi(-2, 6))}")
```

```python
Zi.use_scalar_mult(False)

print(f"{z0 = }")
print(f"{q1 = }")
print(f"{q1 * z0 = }")
print(f"{z0 * q1 = }")
print("---- Check by increasing order of Zi ----")
z0_ord2 = z0.increase_order(2); print(f"{z0_ord2 = }")
print(f"{q1 * z0_ord2 = }")
print(f"{z0_ord2 * q1 = }")
```

```python
print(f"\n{Zi.use_scalar_mult(False) = }")

print(f"    {q1 * 3 = }")
print(f"    {3 * q1 = }")
print(f"{q1 * Zi(3) = }")
print(f"{Zi(3) * q1 = }")

print(f"\n{Zi.use_scalar_mult(True) = }")

print(f"    {q1 * 3 = }")
print(f"    {3 * q1 = }")
print(f"{q1 * Zi(3) = }")
print(f"{Zi(3) * q1 = }")
```
