# Range ("Grid")

Similar to the ``range`` function in Python, but in this case it should generate an iterator for a **grid** of Gaussian integers.

## First, Some Research into Possible Implementations (found code)

``range`` is implemented in C, but I found several examples of how it might be implemented in Python.

### Examples: Range in Python

The following three examples were found on Reddit: https://www.reddit.com/r/learnpython/comments/yc0dbq/how_is_rangen_implemented/

```python
def range1(start, stop, step):
    while start < stop:
        yield start
        start += step

for i in range1(1, 10, 2):
    print(i)
```

"To make it more Python like"

NOTE: Wrt the function definition below, the / symbol is a special marker used to indicate that all parameters to its left are positional-only. This feature was introduced in Python 3.8.

```python
def range2(start, stop=None, step=1, /):
    if step == 0:
        raise ValueError('range2() arg 3 must not be zero')
    if stop is None:
        start, stop = 0, start
    while start < stop if step > 0 else start > stop:
        yield start
        start += step

for i in range2(1, 10, 2):
    print(i)
```

```python
for i in range2(3):
    if i < 20:
        print(i)
    else:
        break
```

"range is not an iterator, though, it's a sequence. A more accurate Python recreation would be this:"

```python
class range3:
    def __init__(self, start, stop=None, step=1, /):
        if step == 0:
            raise ValueError("range3() arg 3 must not be zero")

        if stop is None:
            stop = start
            start = 0

        self.start = start
        self.stop = stop
        self.step = step

    def __len__(self):
         return (self.stop - self.start + abs(self.step) - 1) // self.step

    def __getitem__(self, index):
        if not 0 <= index < len(self):
            raise IndexError("range object index out of range")

        return index*self.step + self.start

    def __iter__(self):
        value = self.start
        while value < self.stop if self.step > 0 else value > self.stop:
            yield value
            value += self.step

for i in range3(1, 10, 2):
    print(i)
```

### Examples: Complex Grid

1. Using a Generator Function

You can define a custom generator to simulate a range for complex numbers. This is useful for moving along a specific vector in the complex plane.

```python
def complex_range(start, stop_modulus, step):
    """
    Iterates from start, adding step until the magnitude 
    of the current number exceeds stop_modulus.
    """
    current = start
    while abs(current) < stop_modulus:
        yield current
        current += step

# Example: Move along the real and imaginary axes simultaneously
for z in complex_range(0 + 0j, 5, 1 + 1j):
    print(z)
# Output: (0+0j), (1+1j), (2+2j), (3+3j)
```

2. Iterating Separately Over Real and Imaginary Parts

If you need a "grid" of complex numbers, you can nest two standard range() functions and combine the components using the complex() constructor. 

```python
# Create a 3x3 grid of complex integers
complex_grid = [complex(re, im) for re in range(3) for im in range(3)]
print(complex_grid)
# Result: [(0+0j), (0+1j), (0+2j), (1+0j), (1+1j), (1+2j), (2+0j), (2+1j), (2+2j)]
```
